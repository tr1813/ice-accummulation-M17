```python
import xarray 
import matplotlib.pyplot as plt
import pyleoclim as pyleo
from scipy.stats import linregress
#https://psl.noaa.gov/thredds/dodsC/Datasets/ncep.reanalysis.derived/surface/air.mon.mean.nc.html
```

```python
monthly = xarray.load_dataset("https://psl.noaa.gov/thredds/dodsC/Datasets/ncep.reanalysis.derived/surface/air.mon.mean.nc")
```

```python

slov1 =  pyleo.Series(value = monthly.air[:,18,5],time = 1970 + monthly.time.astype(float)/(24*365*3600*1e9))
slov2 =  pyleo.Series(value = monthly.air[:,17,5],time = 1970 + monthly.time.astype(float)/(24*365*3600*1e9))
slov3 =  pyleo.Series(value = monthly.air[:,17,6],time = 1970 + monthly.time.astype(float)/(24*365*3600*1e9))
```

```python
reg1 = linregress(slov1.interp(method = 'linear').value,slov1.interp(method = 'linear').time)
reg2 = linregress(slov2.interp(method = 'linear').value,slov2.interp(method = 'linear').time)
reg3 = linregress(slov3.interp(method = 'linear').value,slov3.interp(method = 'linear').time)
```

```python
fig, ax = slov1.plot(mute = True,**{"linewidth":0.4})

slov2.plot(ax = ax,**{"linewidth":0.4})
slov3.plot(ax = ax,**{"linewidth":0.4})


pyleo.showfig(fig)
pyleo.closefig(fig)
```
